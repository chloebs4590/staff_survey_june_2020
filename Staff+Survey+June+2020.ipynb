```python
#select specific columns from staff df
staff_rev = staff[['campus', 'role', 'satisfied_wfh', 'effectiveness_wfh', 'feelings_return_campus', 
                         'greatest_concern_august', 'looking_forward_august', 'covid_personally', 
                        'covid_professionally', 'updates_leadership']] 
staff_rev
```

```python
#finding the unique values and their counts in the 'Role' column
staff_rev['role'].value_counts()
```

```python
#create copy to then set new values in 'Role' column
staff_rev_role = staff_rev.copy()
index_teacher = staff_rev_role[staff_rev_role['role'].isin(['Classroom Teacher','Dance Teacher',
                                                       'Dance/Art/Music Teacher', 'RSP Teacher',
                                                       'On Call Sub, But mainly Sub RSP'])].index
staff_rev_role.loc[index_teacher,'role'] = 'Teacher'
index_asp_ia = staff_rev_role[staff_rev_role['role'].isin(['Instructional Aide','After School Instructor',
                                                       'SPED IA', 'ASP Site Coordinator/ IA'])].index
staff_rev_role.loc[index_asp_ia,'role'] = 'ASP/IA'
index_other = staff_rev_role[staff_rev_role['role'].isin(['Office/Operations Staff','Counselor ',
                                                       'school psych', 'Costume designer'])].index
staff_rev_role.loc[index_other,'role'] = 'Other'
staff_rev_role['role'].value_counts() #there are 87 rows
```

```python
#set the 'role' column as category
staff_rev_role['role'] = staff_rev_role['role'].astype("category")
```

```python
#total number of rows with missing values in staff_rev_role
np.count_nonzero(np.any(staff_rev_role.isnull(), axis=1)) #3 rows with missing data
```

```python
#remove rows that have missing data
staff_rev_role.dropna(axis=0)
```

```python
#finding the unique values and their counts in the 'greatest_concern_august' column
staff_rev_role['greatest_concern_august'].value_counts()
```

```python
#filter out 'other' responses to 'greatest_concern_august'
concerns = ['Personal safety: Being able to remain socially distant from staff and students', 
           'Managing both in-person and remote learning simultaneously',
           'Cleanliness and sanitation',
           'Personal impact on family/living situation',
           'Ability to adhere to changes in school configurations']
staff_rev_role_concern = staff_rev_role[staff_rev_role['greatest_concern_august'].isin(concerns)]
staff_rev_role_concern
staff_rev_role_concern['greatest_concern_august'].value_counts()
```

```python
#finding the unique values and their counts in the 'looking_forward_august' column
staff_rev_role['looking_forward_august'].value_counts()
```

```python
#filter out 'other' responses to 'looking_forward_august'
forward = ['Seeing the students/In-person instruction', 
           'Working alongside colleagues',
           'Increased effectiveness',
           'Daily routines']
staff_rev_role_forward = staff_rev_role[staff_rev_role['looking_forward_august'].isin(forward)]
staff_rev_role_forward
staff_rev_role_forward['looking_forward_august'].value_counts()
```
